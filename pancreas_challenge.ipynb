```python


import os, re, random
from pathlib import Path
from collections import defaultdict, Counter
from typing import Dict, List, Optional, Tuple

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nibabel as nib
import SimpleITK as sitk


ROOT = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/Data/")  
RANDOM_SEED = 7
N_EXAMPLES = 6
AFFINE_ATOL = 1e-3

def load_img(path: Path):
    """Return (array[x,y,z], affine[4x4])."""
    if _backend == "nibabel":
        im = nib.load(str(path))
        return np.asanyarray(im.dataobj), im.affine
    else:
        im = sitk.ReadImage(str(path))
        arr = sitk.GetArrayFromImage(im)         # z,y,x
        arr = np.transpose(arr, (2,1,0))         # x,y,z
        sp = im.GetSpacing() if im.HasSpacing() else (1,1,1)
        aff = np.eye(4, dtype=float)
        aff[0,0], aff[1,1], aff[2,2] = sp[:3]
        return arr, aff

def id_from_image_name(p: Path) -> str:
    """quiz_0_041_0000.nii.gz -> quiz_0_041"""
    s = p.name
    s = re.sub(r"\.nii(\.gz)?$", "", s)
    s = re.sub(r"_0000$", "", s) 
    return s

def id_from_mask_name(p: Path) -> str:
    """quiz_0_041.nii.gz -> quiz_0_041"""
    s = p.name
    s = re.sub(r"\.nii(\.gz)?$", "", s)
    return s

def scan_split(split_dir: Path) -> Tuple[Dict[str, Dict], Dict[str, str]]:
    """
    Return:
      cases: case_id -> dict(image=Path, mask=Path or None, subtype=int)
      problems: map of problems encountered
    """
    cases = {}
    problems = defaultdict(list)
    if split_dir.name in ["train", "validation"]:
        for st in ["subtype0", "subtype1", "subtype2"]:
            d = split_dir / st
            if not d.exists(): 
                continue
            imgs = sorted(d.glob("*_0000.nii*"))
            msks = sorted(d.glob("*.nii*"))
          
            msks = [m for m in msks if not m.name.endswith("_0000.nii.gz") and not m.name.endswith("_0000.nii")]
            
            img_by_id = {id_from_image_name(p): p for p in imgs}
            msk_by_id = {id_from_mask_name(p): p for p in msks}
            for cid, ipath in img_by_id.items():
                mpath = msk_by_id.get(cid, None)
                cases[cid] = {"image": ipath, "mask": mpath, "subtype": int(st[-1])}
            
            for cid in msk_by_id:
                if cid not in img_by_id:
                    problems["extra_mask_no_image"].append(str(d / (cid + ".nii.gz")))
    else:
        
        imgs = sorted(split_dir.glob("*_0000.nii*"))
        for ipath in imgs:
            cid = id_from_image_name(ipath)
            cases[cid] = {"image": ipath, "mask": None, "subtype": None}
    return cases, problems

def header_checks(image_p: Path, mask_p: Optional[Path]) -> List[str]:
    issues = []
    img_arr, img_aff = load_img(image_p)
    ref_shape = img_arr.shape
    if mask_p is not None:
        m_arr, m_aff = load_img(mask_p)
        if m_arr.shape != ref_shape:
            issues.append(f"shape mismatch: {image_p.name} {ref_shape} vs {mask_p.name} {m_arr.shape}")
        if not np.allclose(m_aff, img_aff, atol=AFFINE_ATOL):
            issues.append(f"affine mismatch: {image_p.name} vs {mask_p.name}")
    return issues

def label_sanity(mask_p: Path) -> Tuple[List[int], str]:
    m_arr, _ = load_img(mask_p)
    uniq = np.unique(m_arr).astype(int).tolist()
    if set(uniq).issubset({0,1}):
        kind = "binary"
    elif all(float(v).is_integer() for v in uniq):
        kind = "multi-class"
    else:
        kind = "non-integer labels"
    return uniq, kind

def intensity_stats(img_p: Path) -> dict:
    arr, _ = load_img(img_p)
    arr = arr.astype(np.float32)
    m = np.isfinite(arr)
    vals = arr[m]
    mean = float(vals.mean()) if vals.size else np.nan
    std  = float(vals.std()) if vals.size else np.nan
    vmin = float(vals.min()) if vals.size else np.nan
    vmax = float(vals.max()) if vals.size else np.nan
    p99  = float(np.percentile(vals, 99)) if vals.size else np.nan
    looks_norm = (vmin >= -0.1 and vmax <= 2.0 and std <= 0.6)
    return {"mean":mean, "std":std, "min":vmin, "max":vmax, "p99":p99, "normalized":looks_norm}

def choose_slice(vol: np.ndarray, mask: Optional[np.ndarray]) -> int:
    if mask is not None and mask.ndim == 3:
        z_any = np.where(mask.any(axis=(0,1)))[0]
        if len(z_any) > 0:
            return int(np.median(z_any))
    return vol.shape[2] // 2

def plot_examples(cases: Dict[str, Dict], n: int = 6, title_prefix: str = ""):
    ids = list(cases.keys())
    random.shuffle(ids)
    ids = ids[:min(n, len(ids))]
    for cid in ids:
        ip = cases[cid]["image"]
        mp = cases[cid]["mask"]
        img, _ = load_img(ip)
        msk = None
        if mp is not None:
            msk, _ = load_img(mp)
            msk = (msk > 0).astype(np.uint8)
        z = choose_slice(img, msk)
        plt.figure(figsize=(4,4))
        plt.imshow(img[:,:,z], cmap="gray")
        if msk is not None:
            plt.contour(msk[:,:,z], levels=[0.5], linewidths=1)
        plt.title(f"{title_prefix}{cid} | z={z}")
        plt.axis("off")
        plt.tight_layout()
        plt.show()


train_cases, train_probs = scan_split(ROOT / "train")
val_cases,   val_probs   = scan_split(ROOT / "validation")
test_cases,  test_probs  = scan_split(ROOT / "test")


print("=== Counts ===")
print(f"train: {len(train_cases)}  validation: {len(val_cases)}  test: {len(test_cases)}")
print("train subtype distribution:", Counter([d["subtype"] for d in train_cases.values()]))
print("val   subtype distribution:", Counter([d["subtype"] for d in val_cases.values()]))

issues = []
missing_masks = []
for cid, rec in {**train_cases, **val_cases}.items():
    if rec["mask"] is None:
        missing_masks.append(cid)
    else:
        issues += header_checks(rec["image"], rec["mask"])

if missing_masks:
    print(f"\nMissing masks for {len(missing_masks)} cases (showing first 10):", missing_masks[:10])

print("\nHeader issues (first 15):")
for s in issues[:15]:
    print(" -", s)
if len(issues) > 15:
    print(f"... and {len(issues)-15} more")

label_rows = []
for cid, rec in list(train_cases.items())[:20]:
    mp = rec["mask"]
    if mp is not None:
        uniq, kind = label_sanity(mp)
        label_rows.append({"case_id": cid, "unique_values": uniq, "label_type": kind})
label_df = pd.DataFrame(label_rows)
display(label_df if not label_df.empty else pd.DataFrame({"info": ["No labels to sample"]}))


stat_rows = []
sample_ids = list(train_cases.keys())[:40]
for cid in sample_ids:
    ip = train_cases[cid]["image"]
    st = intensity_stats(ip)
    st.update({"case_id": cid, "file": ip.name})
    stat_rows.append(st)
stats_df = pd.DataFrame(stat_rows)
display(stats_df if not stats_df.empty else pd.DataFrame({"info": ["No images found"]}))

if not stats_df.empty:
    print("\nHeuristic normalization rate:", f"{stats_df['normalized'].mean():.2f}")


print("\nPlotting train examples...")
plot_examples(train_cases, n=N_EXAMPLES, title_prefix="train: ")
print("Plotting validation examples...")
plot_examples(val_cases, n=N_EXAMPLES, title_prefix="val: ")

print("\nDone.")
```

```python
from pathlib import Path
import glob, re


DATA = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/Data/")  

def count_split(split_dir: Path, expect_masks: bool):
    subtypes = sorted([p for p in split_dir.glob("subtype*") if p.is_dir()]) if split_dir.name != "test" else [split_dir]
    report = []
    total_imgs = total_msks = 0
    unpaired = {}

    for st in subtypes:
        if st.name == "test":
            imgs = sorted(st.glob("*_0000.nii.gz"))
            msks = []
        else:
            imgs = sorted(st.glob("*_0000.nii.gz"))
            msks = sorted([p for p in st.glob("*.nii.gz") if not p.name.endswith("_0000.nii.gz")])

        total_imgs += len(imgs)
        total_msks += len(msks)

      
        bad = []
        if expect_masks:
            mset = {m.name for m in msks}
            for im in imgs:
                base = im.name.replace("_0000.nii.gz", "") + ".nii.gz"
                if base not in mset:
                    bad.append(im.name)
            if bad:
                unpaired[st.name] = bad

        report.append((st.name, len(imgs), len(msks)))

    return report, total_imgs, total_msks, unpaired


train_rep, train_imgs, train_msks, train_bad = count_split(DATA/"train", expect_masks=True)
val_rep,   val_imgs,   val_msks,   val_bad   = count_split(DATA/"validation", expect_masks=True)
test_rep,  test_imgs,  test_msks,  test_bad  = count_split(DATA/"test", expect_masks=False)


def pretty_print(title, rep, ti, tm, bad):
    print(f"\n[{title}]")
    for name, ni, nm in rep:
        if name == "test":
            print(f"{name:10s}  images={ni}")
        else:
            print(f"{name:10s}  images={ni:4d}  masks={nm:4d}")
    if title != "test":
        print(f"Total        images={ti:4d}  masks={tm:4d}")
    else:
        print(f"Total        images={ti:4d}")
    if bad:
        for st, lst in bad.items():
            print(f"Unpaired in {st}: {len(lst)} examples (showing up to 10): {lst[:10]}")

pretty_print("train", train_rep, train_imgs, train_msks, train_bad)
pretty_print("validation", val_rep, val_imgs, val_msks, val_bad)
pretty_print("test", test_rep, test_imgs, test_msks, test_bad)


print("\n[Grand totals]")
print(f"Train images={train_imgs}, masks={train_msks}")
print(f"Val   images={val_imgs}, masks={val_msks}")
print(f"Test  images={test_imgs}")
```

#### Classification JSON

```python


import json, re
from pathlib import Path


RAW = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/dataset/nnUNet_raw")
DS  = RAW / "Dataset701_MultiTaskSegClass"
DATASET_JSON = DS / "dataset.json"

PREP = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/dataset/nnUNet_preprocessed")
PREP_DS = PREP / "Dataset701_MultiTaskSegClass"
SPLITS_JSON = PREP_DS / "splits_final.json"          
SUBTYPE_LABELS_JSON = PREP_DS / "subtype_labels.json" 

def infer_class_from_case_id(case_id: str):
    """
    Expected case_id patterns:
      Train/Val: quiz_0_###, quiz_1_###, quiz_2_###  -> class 0/1/2
      Test: quiz_### (no subtype prefix)             -> class None
    """
    m = re.match(r"^quiz_(\d)_", case_id)
    return int(m.group(1)) if m else None

def main():
    assert DATASET_JSON.exists(), f"Missing {DATASET_JSON}"
    with open(DATASET_JSON, "r") as f:
        dsj = json.load(f)


    dsj.setdefault("class_names", {"0": "subtype0", "1": "subtype1", "2": "subtype2"})
    dsj["numClassesClassification"] = 3


    train_entries = dsj.get("training", [])
    case_to_class = {}

  
    for rec in train_entries:
        cid = rec.get("case_id") or Path(rec["image"]).name.replace("_0000.nii.gz", "")
        cls = infer_class_from_case_id(cid)
        if cls is None:
            raise ValueError(f"Cannot infer class from case_id: {cid}")
        case_to_class[cid] = int(cls)
      
        rec["class"] = int(cls)


    if SPLITS_JSON.exists():
        with open(SPLITS_JSON, "r") as f:
            splits = json.load(f)
      
        val_ids = []
        if isinstance(splits, list) and len(splits) > 0:
            val_ids = splits[0].get("val", [])
        for cid in val_ids:
            cls = infer_class_from_case_id(cid)
            if cls is None:
                raise ValueError(f"Cannot infer class from case_id in val: {cid}")
            case_to_class[cid] = int(cls)
    else:
        print(f"Note: {SPLITS_JSON} not found. Skipping val labels.")


    dsj["classification_labels"] = case_to_class

   
    backup_path = DATASET_JSON.with_suffix(".json.bak")
    if not backup_path.exists():
        backup_path.write_text(json.dumps(dsj, indent=2)) 
    DATASET_JSON.write_text(json.dumps(dsj, indent=2))
    print(f"Updated {DATASET_JSON}")


    PREP_DS.mkdir(parents=True, exist_ok=True)
    with open(SUBTYPE_LABELS_JSON, "w") as f:
        json.dump(case_to_class, f, indent=2)
    print(f"Wrote {SUBTYPE_LABELS_JSON} with {len(case_to_class)} labels")

if __name__ == "__main__":
    main()
```

```python
import os, json, csv, sys, numpy as np
from pathlib import Path
import nibabel as nib

try:
    from scipy import ndimage as ndi
except Exception as e:
    raise RuntimeError("This script requires SciPy. Please `pip install scipy`.") from e


RAW = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/dataset/nnUNet_raw")
PREP = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/dataset/nnUNet_preprocessed")
RES  = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/dataset/nnUNet_results")

TASK    = "Dataset701_MultiTaskSegClass"
CFG     = "3d_fullres"
PLANS   = "nnUNetResEncUNetMPlans"
TRAINER = "nnUNetTrainerChallenge"
FOLD    = 0


NSD_TAU_MM = float(os.environ.get("NSD_TAU_MM", "2.0"))


with open(PREP/TASK/"splits_final.json") as f:
    split = json.load(f)[0]
val_ids = split["val"]
val_set = set(val_ids)
print("Validation IDs in split:", len(val_ids))

with open(RAW/TASK/"dataset.json") as f:
    dj = json.load(f)

cid_to_gt = {}
for rec in dj.get("training", []):
    cid = rec.get("case_id")
    if cid in val_set:
        cid_to_gt[cid] = RAW/TASK/rec["label"]

run_root = RES/TASK/f"{TRAINER}__{PLANS}__{CFG}"/f"fold_{FOLD}"
cand_subdirs = [
    "validation_tta_postprocessed",
    "validation_tta",
    "validation_postprocessed",
    "validation",
    "validation_raw",
]
pred_dir = None
for sd in cand_subdirs:
    p = run_root/sd
    if p.exists() and (any(p.glob("*.nii.gz")) or any(p.glob("*.npz"))):
        pred_dir = p
        break

if pred_dir is None:
    cands = []
    for p in run_root.rglob("*"):
        if p.is_dir():
            n = sum(1 for _ in p.glob("*.nii.gz")) + sum(1 for _ in p.glob("*.npz"))
            if n > 0:
                cands.append((n, p))
    cands.sort(reverse=True)
    pred_dir = cands[0][1] if cands else None

if pred_dir is None:
    raise FileNotFoundError(f"No predictions found under {run_root}")

print("Using prediction folder:", pred_dir)


def _ensure_uint_labels(arr):
    if arr.dtype.kind == "f":
        arr = np.rint(arr)
    return arr.astype(np.uint8, copy=False)

def dice_from_bool(a, b):
    inter = np.logical_and(a, b).sum()
    s = a.sum() + b.sum()
    return 1.0 if s == 0 else 2.0 * inter / s

def _get_spacing(img_nii):

    z = img_nii.header.get_zooms()[:3]
    if all(v > 0 for v in z):
        return np.array(z, dtype=float)
    A = img_nii.affine[:3, :3]
    return np.sqrt((A * A).sum(axis=0))

def _surface_mask(bin_mask):
    if not bin_mask.any():
        return np.zeros_like(bin_mask, dtype=bool)
   
    structure = ndi.generate_binary_structure(3, 1)  
    er = ndi.binary_erosion(bin_mask, structure=structure, iterations=1, border_value=0)
    return np.logical_and(bin_mask, np.logical_not(er))

def _distances_between_surfaces(a_bin, b_bin, spacing):
    """
    Return distances from a surface to b mask, and from b surface to a mask (in mm).
    Uses EDT of the complement of the target mask so values are distances to the boundary.
    """
    a_surf = _surface_mask(a_bin)
    b_surf = _surface_mask(b_bin)

    if not a_bin.any() and not b_bin.any():
    
        return np.array([]), np.array([])

    
    b_inv = np.logical_not(b_bin)
    a_inv = np.logical_not(a_bin)
    dt_b = ndi.distance_transform_edt(b_inv, sampling=spacing)
    dt_a = ndi.distance_transform_edt(a_inv, sampling=spacing)

    da = dt_b[a_surf]  
    db = dt_a[b_surf]  
    return da, db

def hd95_msd_nsd(a_bin, b_bin, spacing, tau_mm):
    """
    Returns HD95 (mm), MSD (mm), NSD (%) using symmetric definition.
    Handles empty-set edge cases as described in the notes.
    """
    
    if not a_bin.any() and not b_bin.any():
        return 0.0, 0.0, 100.0

    
    if a_bin.any() != b_bin.any():
    
        return np.nan, np.nan, 0.0

    da, db = _distances_between_surfaces(a_bin, b_bin, spacing)
    if da.size + db.size == 0:
    
        return np.nan, np.nan, 0.0

    both = np.concatenate([da, db])
    hd95 = float(np.percentile(both, 95))
    msd = float(both.mean())

    
    within = (da <= tau_mm).sum() + (db <= tau_mm).sum()
    total = da.size + db.size
    nsd = 100.0 * float(within) / float(total) if total > 0 else 0.0
    return hd95, msd, nsd

def load_pred_for_case(cid):
    """
    Try common filenames for either .nii.gz or .npz in pred_dir.
    Returns np.ndarray label volume.
    """
    
    nii_names = [
        f"{cid}.nii.gz",
        f"{cid}_seg.nii.gz",
        f"{cid}_0000.nii.gz",
        f"{cid}_0000_seg.nii.gz",
    ]
    for name in nii_names:
        f = pred_dir / name
        if f.exists():
            img = nib.load(str(f))
            arr = img.get_fdata()
            return _ensure_uint_labels(arr)

    
    npz_names = [
        f"{cid}.npz",
        f"{cid}_seg.npz",
        f"{cid}_0000.npz",
        f"{cid}_0000_seg.npz",
    ]
    for name in npz_names:
        f = pred_dir / name
        if f.exists():
            d = np.load(f)
            arr = None
            for k in ["softmax", "probabilities", "pred"]:
                if k in d:
                    arr = d[k]
                    break
            if arr is None:
                
                for k in d.files:
                    if isinstance(d[k], np.ndarray):
                        arr = d[k]
                        break
            if arr is None or arr.ndim != 4:
                raise RuntimeError(f"Unexpected npz contents in {f}")
            lab = np.argmax(arr, axis=0).astype(np.uint8)
            return lab

    return None



rows = []
missing_pred, missing_gt, shape_mismatch = [], [], []


agg = {
    "whole": {"dice": [], "hd95": [], "msd": [], "nsd": []},
    "lesion": {"dice": [], "hd95": [], "msd": [], "nsd": []},
}

for cid in val_ids:
    gt_p = cid_to_gt.get(cid)
    if gt_p is None or not Path(gt_p).exists():
        missing_gt.append(cid)
        continue

    pred = load_pred_for_case(cid)
    if pred is None:
        missing_pred.append(cid)
        continue

    gt_img = nib.load(str(gt_p))
    gt = _ensure_uint_labels(gt_img.get_fdata())
    if gt.shape != pred.shape:
        shape_mismatch.append((cid, gt.shape, pred.shape))
        continue

    spacing = np.array(_get_spacing(gt_img), dtype=float)


    gt_whole = np.logical_or(gt == 1, gt == 2)
    pr_whole = np.logical_or(pred == 1, pred == 2)
    gt_les   = (gt == 2)
    pr_les   = (pred == 2)


    d_whole = float(dice_from_bool(pr_whole, gt_whole))
    d_les   = float(dice_from_bool(pr_les,   gt_les))

    hd_whole, msd_whole, nsd_whole = hd95_msd_nsd(pr_whole, gt_whole, spacing, NSD_TAU_MM)
    hd_les,   msd_les,   nsd_les   = hd95_msd_nsd(pr_les,   gt_les,   spacing, NSD_TAU_MM)

    rows.append((
        cid,
        d_whole, d_les,
        hd_whole, msd_whole, nsd_whole,
        hd_les,   msd_les,   nsd_les
    ))


    agg["whole"]["dice"].append(d_whole)
    agg["lesion"]["dice"].append(d_les)
    for k, v in zip(("hd95","msd","nsd"), (hd_whole, msd_whole, nsd_whole)):
        if not np.isnan(v):
            agg["whole"][k].append(float(v))
    for k, v in zip(("hd95","msd","nsd"), (hd_les, msd_les, nsd_les)):
        if not np.isnan(v):
            agg["lesion"][k].append(float(v))

print("Evaluated cases:", len(rows))

def _mean_sd_fmt(vals, unit=""):
    if len(vals) == 0:
        return "n/a"
    m = np.mean(vals)
    s = np.std(vals, ddof=1) if len(vals) > 1 else 0.0
    if unit == "%":
        return f"{m:.1f}  {s:.1f} {unit}"
    return f"{m:.3f}  {s:.3f} {unit}".strip()


print(f"NSD tolerance: {NSD_TAU_MM:.2f} mm\n")

print("Whole pancreas (labels 1|2)")
print("  Dice:", _mean_sd_fmt(agg["whole"]["dice"]))
print("  MSD:",  _mean_sd_fmt(agg["whole"]["msd"],  "mm"))
print("  HD95:", _mean_sd_fmt(agg["whole"]["hd95"], "mm"))
print("  NSD:",  _mean_sd_fmt(agg["whole"]["nsd"],  "%"))

print("\nLesion only (label 2)")
print("  Dice:", _mean_sd_fmt(agg["lesion"]["dice"]))
print("  MSD:",  _mean_sd_fmt(agg["lesion"]["msd"],  "mm"))
print("  HD95:", _mean_sd_fmt(agg["lesion"]["hd95"], "mm"))
print("  NSD:",  _mean_sd_fmt(agg["lesion"]["nsd"],  "%"))


def _macro(vals_a, vals_b):
    a = np.array(vals_a, dtype=float)
    b = np.array(vals_b, dtype=float)
    if a.size == 0 and b.size == 0:
        return []

    return list(a) + list(b)

print("\nCombined (macro over whole and lesion)")
print("  Dice:", _mean_sd_fmt(_macro(agg["whole"]["dice"], agg["lesion"]["dice"])))
print("  MSD:",  _mean_sd_fmt(_macro(agg["whole"]["msd"],  agg["lesion"]["msd"]),  "mm"))
print("  HD95:", _mean_sd_fmt(_macro(agg["whole"]["hd95"], agg["lesion"]["hd95"]), "mm"))
print("  NSD:",  _mean_sd_fmt(_macro(agg["whole"]["nsd"],  agg["lesion"]["nsd"]),  "%"))

if missing_pred:
    print(f"[missing pred] {len(missing_pred)} cases. Sample:", missing_pred[:8])
if missing_gt:
    print(f"[missing GT]   {len(missing_gt)} cases. Sample:", missing_gt[:8])
if shape_mismatch:
    print(f"[shape mismatch] {len(shape_mismatch)} cases. Sample:", shape_mismatch[:3])


out_csv = pred_dir/"val_metrics_summary.csv"
with open(out_csv, "w", newline="") as f:
    w = csv.writer(f)
    w.writerow([
        "case_id",
        "dice_whole_pancreas_1or2",
        "dice_lesion_2",
        "hd95_whole_mm", "msd_whole_mm", "nsd_whole_percent",
        "hd95_lesion_mm", "msd_lesion_mm", "nsd_lesion_percent",
        "nsd_tau_mm"
    ])
    for r in rows:
        cid, d_w, d_l, hd_w, msd_w, nsd_w, hd_l, msd_l, nsd_l = r
        w.writerow([
            cid, f"{d_w:.4f}", f"{d_l:.4f}",
            f"{hd_w:.4f}" if not np.isnan(hd_w) else "nan",
            f"{msd_w:.4f}" if not np.isnan(msd_w) else "nan",
            f"{nsd_w:.2f}",
            f"{hd_l:.4f}" if not np.isnan(hd_l) else "nan",
            f"{msd_l:.4f}" if not np.isnan(msd_l) else "nan",
            f"{nsd_l:.2f}",
            f"{NSD_TAU_MM:.2f}",
        ])
print("Per-case CSV:", out_csv)
```

```python


import os, time, json, csv, types, sys, inspect
from pathlib import Path
import numpy as np
import nibabel as nib
import torch


RAW = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/dataset/nnUNet_raw")
PREP = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/dataset/nnUNet_preprocessed")
RES  = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/dataset/nnUNet_results")

TASK    = "Dataset701_MultiTaskSegClass"
CFG     = "3d_fullres"
PLANS   = "nnUNetResEncUNetMPlans"
TRAINER = "nnUNetTrainerChallenge"
FOLD    = 0

OUT_DIR = Path("/tmp/fast_preds_v2")
DISABLE_TTA = True
NSD_TAU_MM = float(os.environ.get("NSD_TAU_MM", "2.0"))


stub_name = "nnunetv2.training.nnUNetTrainer.fast_predict_and_eval"
if stub_name not in sys.modules:
    sys.modules[stub_name] = types.ModuleType(stub_name)


from nnunetv2.inference.predict_from_raw_data import nnUNetPredictor
from nnunetv2.imageio.simpleitk_reader_writer import SimpleITKIO


try:
    from scipy import ndimage as ndi
except Exception as e:
    raise RuntimeError("SciPy is required. pip install scipy") from e

def _ensure_uint_labels(arr: np.ndarray) -> np.ndarray:
    if arr.dtype.kind == "f":
        arr = np.rint(arr)
    return arr.astype(np.uint8, copy=False)

def dice_from_bool(a: np.ndarray, b: np.ndarray) -> float:
    inter = np.logical_and(a, b).sum()
    s = a.sum() + b.sum()
    return 1.0 if s == 0 else 2.0 * inter / s

def _get_spacing(img_nii: nib.Nifti1Image) -> np.ndarray:
    z = img_nii.header.get_zooms()[:3]
    if all(v > 0 for v in z):
        return np.array(z, dtype=float)
    A = img_nii.affine[:3, :3]
    return np.sqrt((A * A).sum(axis=0))

def _surface_mask(bin_mask: np.ndarray) -> np.ndarray:
    if not bin_mask.any():
        return np.zeros_like(bin_mask, dtype=bool)
    structure = ndi.generate_binary_structure(3, 1)
    er = ndi.binary_erosion(bin_mask, structure=structure, iterations=1, border_value=0)
    return np.logical_and(bin_mask, np.logical_not(er))

def _distances_between_surfaces(a_bin: np.ndarray, b_bin: np.ndarray, spacing: np.ndarray):
    a_surf = _surface_mask(a_bin)
    b_surf = _surface_mask(b_bin)
    if not a_bin.any() and not b_bin.any():
        return np.array([]), np.array([])
    b_inv = np.logical_not(b_bin)
    a_inv = np.logical_not(a_bin)
    dt_b = ndi.distance_transform_edt(b_inv, sampling=spacing)
    dt_a = ndi.distance_transform_edt(a_inv, sampling=spacing)
    da = dt_b[a_surf]
    db = dt_a[b_surf]
    return da, db

def hd95_msd_nsd(pred_bin: np.ndarray, gt_bin: np.ndarray, spacing: np.ndarray, tau_mm: float):
    if not pred_bin.any() and not gt_bin.any():
        return 0.0, 0.0, 100.0
    if pred_bin.any() != gt_bin.any():
        return np.nan, np.nan, 0.0
    da, db = _distances_between_surfaces(pred_bin, gt_bin, spacing)
    if da.size + db.size == 0:
        return np.nan, np.nan, 0.0
    both = np.concatenate([da, db])
    hd95 = float(np.percentile(both, 95))
    msd = float(both.mean())
    within = (da <= tau_mm).sum() + (db <= tau_mm).sum()
    total = da.size + db.size
    nsd = 100.0 * float(within) / float(total) if total > 0 else 0.0
    return hd95, msd, nsd

def _mean_sd_fmt(vals, unit=""):
    if len(vals) == 0:
        return "n/a"
    m = np.mean(vals)
    s = np.std(vals, ddof=1) if len(vals) > 1 else 0.0
    if unit == "%":
        return f"{m:.1f}  {s:.1f} {unit}"
    return f"{m:.3f}  {s:.3f} {unit}".strip()

def _macro(*lists):
    out = []
    for L in lists:
        out.extend(list(L))
    return out


def _amp_dtype():
    if torch.cuda.is_available():
        if torch.cuda.is_bf16_supported():
            return torch.bfloat16
        return torch.float16
    return torch.float32

def _pick_checkpoint_name(fold_dir: Path) -> str:
    for name in ("checkpoint_final.pth", "checkpoint_best.pth"):
        if (fold_dir / name).exists():
            return name
    cand = list(fold_dir.glob("*.pth"))
    if cand:
        return cand[0].name
    raise FileNotFoundError(f"No checkpoint found in {fold_dir}")

def fast_predict_folder(
    input_ids,
    raw_task_dir: Path,
    out_dir: Path,
    model_base_dir: Path,
    folds=(0,),
    disable_tta=True,
    verbose=False
):
    out_dir.mkdir(parents=True, exist_ok=True)


    torch.backends.cudnn.benchmark = True
    torch.backends.cuda.matmul.allow_tf32 = True
    torch.backends.cudnn.allow_tf32 = True
    torch.set_float32_matmul_precision("high")
    amp_dtype = _amp_dtype()
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


    first_fold = int(folds[0]) if isinstance(folds, (list, tuple)) and len(folds) else 0
    ckpt_name = _pick_checkpoint_name(model_base_dir / f"fold_{first_fold}")

    predictor = nnUNetPredictor(
        tile_step_size=0.5,
        use_gaussian=True,
        use_mirroring=not disable_tta,
        verbose=verbose,
        verbose_preprocessing=False,
        allow_tqdm=False,
        device=device
    )
    if hasattr(predictor, "perform_everything_on_gpu"):
        try:
            predictor.perform_everything_on_gpu = True
        except Exception:
            pass

    init_fn = predictor.initialize_from_trained_model_folder
    params = list(inspect.signature(init_fn).parameters.keys())
    try:
        if "use_folds" in params:
            init_fn(str(model_base_dir), use_folds=folds, checkpoint_name=ckpt_name)
        elif "folds" in params:
            init_fn(str(model_base_dir), folds=folds, checkpoint_name=ckpt_name)
        else:
            init_fn(str(model_base_dir), checkpoint_name=ckpt_name)
            if hasattr(predictor, "folds"):
                predictor.folds = list(folds)
    except TypeError:
        init_fn(str(model_base_dir), checkpoint_name=ckpt_name)
        if hasattr(predictor, "folds"):
            predictor.folds = list(folds)

    net = predictor.network
    net.eval().to(device)


    for m in net.modules():
        if isinstance(m, torch.nn.Conv3d):
            try:
                m.to(memory_format=torch.channels_last_3d)
            except Exception:
                pass

    reader = SimpleITKIO()
    times = []
    with torch.inference_mode():
        for cid in input_ids:
            hits = sorted((raw_task_dir / "imagesTr").glob(f"{cid}_*.nii.gz")) + \
                   sorted((raw_task_dir / "imagesTs").glob(f"{cid}_*.nii.gz"))
            if not hits:
                print(f"Skip {cid}: no image files in {raw_task_dir}/imagesTr or imagesTs")
                continue
            img_list = [str(p) for p in sorted(hits, key=lambda p: p.name)]
            arrays, props = reader.read_images(img_list)

            t0 = time.time()
            
            with torch.autocast(device_type="cuda", dtype=amp_dtype, enabled=torch.cuda.is_available()):
                try:
                    out = predictor.predict_single_npy_array(arrays, props, seg_from_prev_stage=None)
                except TypeError:
                    out = predictor.predict_single_npy_array(arrays, props)
        
            if isinstance(out, tuple):
                pred_label = out[0]
            else:
                pred_label = out
            times.append(time.time() - t0)
            
            gt_hits = [RAW / TASK / rec["label"] for rec in json.load(open(RAW / TASK / "dataset.json")).get("training", [])
                    if rec.get("case_id") == cid]
            if gt_hits and gt_hits[0].exists():
                gt_shape = nib.load(str(gt_hits[0])).shape
                if len(gt_shape) == 3 and len(pred_label.shape) == 3:
                    if gt_shape == (pred_label.shape[2], pred_label.shape[1], pred_label.shape[0]):
                        pred_label = np.transpose(pred_label, (2, 1, 0))

            nib.save(nib.Nifti1Image(pred_label.astype(np.uint8), np.eye(4)), str(out_dir / f"{cid}.nii.gz"))

    mean_t = float(np.mean(times)) if times else 0.0
    print(f"[fast inference] {len(times)} cases. Mean {mean_t:.2f} s per case.")
    return times


def evaluate_split(task_name: str, fold: int, pred_dir: Path, nsd_tau_mm: float):
    split = json.load(open(PREP / task_name / "splits_final.json"))[fold]
    val_ids = split["val"]
    val_set = set(val_ids)

    dj = json.load(open(RAW / task_name / "dataset.json"))
    cid_to_gt = {}
    for rec in dj.get("training", []):
        cid = rec.get("case_id")
        if cid in val_set:
            cid_to_gt[cid] = RAW / task_name / rec["label"]

    rows = []
    agg = {k: {"dice": [], "hd95": [], "msd": [], "nsd": []} for k in ("whole","lesion","label1")}

    for cid in val_ids:
        gt_p = cid_to_gt.get(cid)
        pred_p = pred_dir / f"{cid}.nii.gz"
        if gt_p is None or not Path(gt_p).exists() or not pred_p.exists():
            continue

        gt_img = nib.load(str(gt_p))
        pred_img = nib.load(str(pred_p))
        gt = _ensure_uint_labels(gt_img.get_fdata(dtype=np.float32))
        pred = _ensure_uint_labels(pred_img.get_fdata(dtype=np.float32))
        if gt.shape != pred.shape:
            print(f"[shape mismatch] {cid} gt {gt.shape} pred {pred.shape}")
            continue

        spacing = _get_spacing(gt_img)
        gt_whole = np.logical_or(gt == 1, gt == 2); pr_whole = np.logical_or(pred == 1, pred == 2)
        gt_les   = (gt == 2); pr_les = (pred == 2)
        gt_lbl1  = (gt == 1); pr_lbl1 = (pred == 1)

        d_whole = float(dice_from_bool(pr_whole, gt_whole))
        d_les   = float(dice_from_bool(pr_les,   gt_les))
        d_lbl1  = float(dice_from_bool(pr_lbl1,  gt_lbl1))

        hd_w, msd_w, nsd_w = hd95_msd_nsd(pr_whole, gt_whole, spacing, nsd_tau_mm)
        hd_l, msd_l, nsd_l = hd95_msd_nsd(pr_les,   gt_les,   spacing, nsd_tau_mm)
        hd_1, msd_1, nsd_1 = hd95_msd_nsd(pr_lbl1,  gt_lbl1,  spacing, nsd_tau_mm)

        rows.append((cid, d_whole, d_les, d_lbl1, hd_w, msd_w, nsd_w, hd_l, msd_l, nsd_l, hd_1, msd_1, nsd_1))

        agg["whole"]["dice"].append(d_whole); agg["lesion"]["dice"].append(d_les); agg["label1"]["dice"].append(d_lbl1)
        for k, v in zip(("hd95","msd","nsd"), (hd_w, msd_w, nsd_w)):
            if not np.isnan(v): agg["whole"][k].append(float(v))
        for k, v in zip(("hd95","msd","nsd"), (hd_l, msd_l, nsd_l)):
            if not np.isnan(v): agg["lesion"][k].append(float(v))
        for k, v in zip(("hd95","msd","nsd"), (hd_1, msd_1, nsd_1)):
            if not np.isnan(v): agg["label1"][k].append(float(v))

    print("Evaluated cases:", len(rows))
    print("\nWhole pancreas (labels 1|2)")
    print("  Dice:", _mean_sd_fmt(agg["whole"]["dice"]))
    print("  MSD:",  _mean_sd_fmt(agg["whole"]["msd"],  "mm"))
    print("  HD95:", _mean_sd_fmt(agg["whole"]["hd95"], "mm"))
    print("  NSD:",  _mean_sd_fmt(agg["whole"]["nsd"],  "%"))

    print("\nLesion only (label 2)")
    print("  Dice:", _mean_sd_fmt(agg["lesion"]["dice"]))
    print("  MSD:",  _mean_sd_fmt(agg["lesion"]["msd"],  "mm"))
    print("  HD95:", _mean_sd_fmt(agg["lesion"]["hd95"], "mm"))
    print("  NSD:",  _mean_sd_fmt(agg["lesion"]["nsd"],  "%"))

    print("\nLabel 1 only")
    print("  Dice:", _mean_sd_fmt(agg["label1"]["dice"]))
    print("  MSD:",  _mean_sd_fmt(agg["label1"]["msd"],  "mm"))
    print("  HD95:", _mean_sd_fmt(agg["label1"]["hd95"], "mm"))
    print("  NSD:",  _mean_sd_fmt(agg["label1"]["nsd"],  "%"))

    print("\nCombined (macro over whole, lesion, label 1)")
    print("  Dice:", _mean_sd_fmt(_macro(agg["whole"]["dice"],  agg["lesion"]["dice"],  agg["label1"]["dice"])))
    print("  MSD:",  _mean_sd_fmt(_macro(agg["whole"]["msd"],   agg["lesion"]["msd"],   agg["label1"]["msd"]),  "mm"))
    print("  HD95:", _mean_sd_fmt(_macro(agg["whole"]["hd95"],  agg["lesion"]["hd95"],  agg["label1"]["hd95"]), "mm"))
    print("  NSD:",  _mean_sd_fmt(_macro(agg["whole"]["nsd"],   agg["lesion"]["nsd"],   agg["label1"]["nsd"]),  "%"))

    out_csv = pred_dir / "val_metrics_summary.csv"
    with open(out_csv, "w", newline="") as f:
        w = csv.writer(f)
        w.writerow([
            "case_id",
            "dice_whole_pancreas_1or2",
            "dice_lesion_2",
            "dice_label1_only",
            "hd95_whole_mm", "msd_whole_mm", "nsd_whole_percent",
            "hd95_lesion_mm", "msd_lesion_mm", "nsd_lesion_percent",
            "hd95_label1_mm", "msd_label1_mm", "nsd_label1_percent",
            "nsd_tau_mm"
        ])
        for r in rows:
            cid, d_w, d_l, d_1, hd_w, msd_w, nsd_w, hd_l, msd_l, nsd_l, hd_1, msd_1, nsd_1 = r
            w.writerow([
                cid,
                f"{d_w:.4f}", f"{d_l:.4f}", f"{d_1:.4f}",
                f"{hd_w:.4f}" if not np.isnan(hd_w) else "nan",
                f"{msd_w:.4f}" if not np.isnan(msd_w) else "nan",
                f"{nsd_w:.2f}",
                f"{hd_l:.4f}" if not np.isnan(hd_l) else "nan",
                f"{msd_l:.4f}" if not np.isnan(msd_l) else "nan",
                f"{nsd_l:.2f}",
                f"{hd_1:.4f}" if not np.isnan(hd_1) else "nan",
                f"{msd_1:.4f}" if not np.isnan(msd_1) else "nan",
                f"{nsd_1:.2f}",
                f"{NSD_TAU_MM:.2f}",
            ])
    print("Per-case CSV:", out_csv)


split = json.load(open(PREP / TASK / "splits_final.json"))[FOLD]
val_ids = split["val"]
print("Validation IDs in split:", len(val_ids))

OUT_DIR.mkdir(parents=True, exist_ok=True)
model_base_dir = RES / TASK / f"{TRAINER}__{PLANS}__{CFG}"

_ = fast_predict_folder(
    input_ids=val_ids,
    raw_task_dir=RAW / TASK,
    out_dir=OUT_DIR,
    model_base_dir=model_base_dir,
    folds=(FOLD,),
    disable_tta=DISABLE_TTA,
    verbose=False
)

evaluate_split(TASK, FOLD, OUT_DIR, nsd_tau_mm=NSD_TAU_MM)
```

```python

import os, re, json
from pathlib import Path
from typing import Dict, List, Tuple
import numpy as np
import nibabel as nib
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D


RAW = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/dataset/nnUNet_raw")
PREP = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/dataset/nnUNet_preprocessed")
RES  = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/dataset/nnUNet_results")

TASK    = "Dataset701_MultiTaskSegClass"
CFG     = "3d_fullres"
PLANS   = "nnUNetResEncUNetMPlans"
TRAINER = "nnUNetTrainerChallenge"
FOLD    = 0

OUT_DIR = Path("/fast_preds_v2")  #
TOP_K = 4
BOTTOM_K = 4
OUT_EXAMPLES = RES / TASK / f"{TRAINER}__{PLANS}__{CFG}" / "valid_examples"


def _ensure_uint_labels(arr: np.ndarray) -> np.ndarray:
    if arr.dtype.kind == "f":
        arr = np.rint(arr)
    return arr.astype(np.uint8, copy=False)

def dice_from_bool(a: np.ndarray, b: np.ndarray) -> float:
    inter = np.logical_and(a, b).sum()
    s = a.sum() + b.sum()
    return 1.0 if s == 0 else 2.0 * inter / s

def _im_norm(slice2d: np.ndarray) -> np.ndarray:
    x = slice2d.astype(np.float32)
    lo, hi = np.percentile(x, 0.5), np.percentile(x, 99.5)
    if hi <= lo:
        hi = lo + 1.0
    return np.clip((x - lo) / (hi - lo), 0, 1)

def _best_lesion_slice(mask3d: np.ndarray) -> int:
    areas = mask3d.sum(axis=(0, 1)).astype(np.int64)
    return int(areas.argmax()) if areas.max() > 0 else int(mask3d.shape[2] // 2)

def _case_score_lesion_dice(gt_lbl: np.ndarray, pr_lbl: np.ndarray) -> float:
    return float(dice_from_bool(pr_lbl == 2, gt_lbl == 2))

def _cid_from_stem(stem: str) -> str:
    s = stem
    if s.lower().endswith(".nii"): 
        s = s[:-4]
    s = re.sub(r"_(0000|0001|0002|0003)$", "", s)
    s = re.sub(r"(_pred|_seg|_mask)$", "", s, flags=re.I)
    return s

def _find_backdrop_image(raw_task_dir: Path, cid: str) -> Path:
    candidates = []
    patt = [f"{cid}_*0000.nii.gz", f"{cid}_*.nii.gz", f"{cid}.nii.gz"]
    roots = [
        raw_task_dir / "imagesTr",
        raw_task_dir / "imagesTs",
        raw_task_dir / "imagesVal",
        raw_task_dir / "labelsVal" / "imagesVal",
    ]
    for root in roots:
        if not root.exists():
            continue
        for p in patt:
            hits = sorted(root.glob(p))
            if hits:
                candidates.extend(hits)
                break
        if candidates:
            break
    if not candidates:
        raise FileNotFoundError(f"No image found for {cid} in imagesTr/imagesTs/imagesVal/labelsVal/imagesVal")
    return candidates[0]

def _discover_predictions(pred_dir: Path, model_base_dir: Path, fold: int) -> Dict[str, Path]:
    pool = []
    if pred_dir.exists():
        pool += list(sorted(pred_dir.rglob("*.nii.gz")))
    val_dir = model_base_dir / f"fold_{fold}" / "validation"
    if val_dir.exists():
        pool += list(sorted(val_dir.rglob("*.nii.gz")))
    uniq = []
    seen = set()
    for p in pool:
        if p not in seen:
            uniq.append(p); seen.add(p)
    by_cid = {}
    for p in uniq:
        stem = p.stem
        if stem.lower().endswith(".nii"):
            stem = stem[:-4]
        cid = _cid_from_stem(stem)
        by_cid[cid] = p
    return by_cid


def save_good_bad_examples_valsplit(
    task_name: str,
    raw_dir: Path,
    prep_dir: Path,
    res_dir: Path,
    pred_dir: Path,
    trainer: str,
    plans: str,
    cfg: str,
    fold: int,
    out_examples: Path,
    top_k: int,
    bottom_k: int,
) -> None:
    out_examples.mkdir(parents=True, exist_ok=True)
    raw_task = raw_dir / task_name
    prep_task = prep_dir / task_name

    split = json.load(open(prep_task / "splits_final.json"))[fold]
    val_ids = list(split["val"])
    print(f"[split] validation IDs: {len(val_ids)}")

    gt_root = prep_task / "gt_segmentations"
    gt_map: Dict[str, Path] = {}
    miss_gt = []
    for cid in val_ids:
        p = gt_root / f"{cid}.nii.gz"
        if p.exists():
            gt_map[cid] = p
        else:
            miss_gt.append(cid)
    if miss_gt:
        print(f"[warn] missing GT for {len(miss_gt)} val IDs in gt_segmentations. Example: {miss_gt[:5]}")

   
    model_base_dir = res_dir / task_name / f"{trainer}__{plans}__{cfg}"
    preds_by_cid = _discover_predictions(pred_dir, model_base_dir, fold)
    print(f"[discover] predictions found: {len(preds_by_cid)}")

 
    overlap = [cid for cid in val_ids if cid in gt_map and cid in preds_by_cid]
    print(f"[match] overlap valgtpred: {len(overlap)}")
    if not overlap:
        print("[hint] sample prediction cids:", list(preds_by_cid.keys())[:10])
        print("[hint] ensure prediction filenames match val ids")
        return

    per_case = []
    for cid in overlap:
        gt_p = gt_map[cid]
        pr_p = preds_by_cid[cid]
        gt = _ensure_uint_labels(nib.load(str(gt_p)).get_fdata(dtype=np.float32))
        pr = _ensure_uint_labels(nib.load(str(pr_p)).get_fdata(dtype=np.float32))
        if gt.shape != pr.shape and gt.shape == (pr.shape[2], pr.shape[1], pr.shape[0]):
            pr = np.transpose(pr, (2, 1, 0))
        if gt.shape != pr.shape:
            print(f"[skip] shape mismatch {cid} gt{gt.shape} pr{pr.shape}")
            continue
        lesion_d = _case_score_lesion_dice(gt, pr)
        whole_d  = float(dice_from_bool(gt > 0, pr > 0))
        per_case.append((cid, lesion_d, whole_d))

    if not per_case:
        print("Matched IDs exist but nothing scored. Check shapes.")
        return

    per_case.sort(key=lambda x: x[1], reverse=True)
    good = per_case[:top_k]
    nonzero_sorted_asc = sorted([r for r in per_case if r[1] > 0.0], key=lambda x: x[1])
    if len(nonzero_sorted_asc) == 0:
        print("[warn] all lesion Dice are 0.0; no bad examples will be saved by request.")
        bad = []
    else:
        bad = nonzero_sorted_asc[:bottom_k]
        if len(bad) < bottom_k:
            print(f"[info] only {len(bad)} non-zero bad examples available; saving those.")


    LOG_ROWS = []  

    
    legend_handles = [
        Line2D([0], [0], color="c", lw=2.0, ls="--", label="GT label1 (cyan, dashed)"),
        Line2D([0], [0], color="r", lw=2.0, ls="--", label="GT label2 (red, dashed)"),
        Line2D([0], [0], color="g", lw=2.0, ls="-",  label="Pred label1 (green, solid)"),
        Line2D([0], [0], color="m", lw=2.0, ls="-",  label="Pred label2 (magenta, solid)"),
    ]

    def _draw_case(cid: str, rank_tag: str, lesion_d: float, whole_d: float):
        gt_p = gt_map[cid]
        pr_p = preds_by_cid[cid]

 
        try:
            img_p = _find_backdrop_image(raw_task, cid)
            img = nib.load(str(img_p)).get_fdata(dtype=np.float32)
        except Exception as e:
            print(f"[warn] no backdrop for {cid}: {e}. Using GT as faint background.")
            img_p = None
            img = _ensure_uint_labels(nib.load(str(gt_p)).get_fdata(dtype=np.float32)).astype(np.float32)

        gt = _ensure_uint_labels(nib.load(str(gt_p)).get_fdata(dtype=np.float32))
        pr = _ensure_uint_labels(nib.load(str(pr_p)).get_fdata(dtype=np.float32))
        if gt.shape != pr.shape and gt.shape == (pr.shape[2], pr.shape[1], pr.shape[0]):
            pr = np.transpose(pr, (2, 1, 0))
        if img.shape != gt.shape and img.shape == (gt.shape[2], gt.shape[1], gt.shape[0]):
            img = np.transpose(img, (2, 1, 0))

        z = _best_lesion_slice(gt == 2)
        z = min(z, img.shape[2] - 1)

        im2d = _im_norm(img[:, :, z])
        gt2, pr2 = gt[:, :, z], pr[:, :, z]

        gt_l1, gt_l2 = (gt2 == 1), (gt2 == 2)
        pr_l1, pr_l2 = (pr2 == 1), (pr2 == 2)

        h, w = im2d.shape
        dpi = 120
        fig = plt.figure(figsize=(max(6, w / 100), max(6.8, h / 100 + 0.8)), dpi=dpi)
        gs = fig.add_gridspec(nrows=3, ncols=1, height_ratios=[1.0, 0.12, 0.20])
        ax = fig.add_subplot(gs[0])
        cap = fig.add_subplot(gs[1]); cap.axis("off")
        leg = fig.add_subplot(gs[2]); leg.axis("off")

        ax.imshow(im2d, cmap="gray", interpolation="nearest")

        ax.contour(gt_l1, levels=[0.5], colors=["c"], linewidths=2.0, linestyles="dashed")
        ax.contour(gt_l2, levels=[0.5], colors=["r"], linewidths=2.0, linestyles="dashed")
    
        ax.contour(pr_l1, levels=[0.5], colors=["g"], linewidths=2.0, linestyles="solid")
        ax.contour(pr_l2, levels=[0.5], colors=["m"], linewidths=2.0, linestyles="solid")
        ax.set_axis_off()

        fig.suptitle(f"{cid} | z={z}", fontsize=12, y=0.985)
        cap_text = f"Lesion Dice = {lesion_d:.3f}   Whole Dice = {whole_d:.3f}"
        cap.text(0.01, 0.5, cap_text, fontsize=11, ha="left", va="center", transform=cap.transAxes)

 
        leg.legend(handles=legend_handles, loc="center", ncol=2, frameon=False)

        save_path = out_examples / f"{rank_tag}_{cid}_z{z:03d}.png"
        plt.savefig(save_path, bbox_inches="tight", pad_inches=0.05)
        plt.close(fig)

        LOG_ROWS.append([
            rank_tag, cid, z, f"{lesion_d:.4f}", f"{whole_d:.4f}",
            str(pr_p), str(gt_p), str(img_p) if img_p is not None else ""
        ])

    for i, (cid, les_d, whole_d) in enumerate(good, start=1):
        _draw_case(cid, f"good_{i:02d}", les_d, whole_d)
    for i, (cid, les_d, whole_d) in enumerate(bad, start=1):
        _draw_case(cid, f"bad_{i:02d}", les_d, whole_d)


    scores_txt = out_examples / "examples_scores.tsv"
    with open(scores_txt, "w") as f:
        f.write("rank_tag\tcase_id\tz\tlesion_dice\twhole_dice\tpred_path\tgt_path\timage_path\n")
        for row in LOG_ROWS:
            f.write("\t".join(map(str, row)) + "\n")

    print(f"Wrote {len(good)} good and {len(bad)} bad images to: {out_examples}")
    print(f"Scores saved to: {scores_txt}")


save_good_bad_examples_valsplit(
    task_name=TASK,
    raw_dir=RAW,
    prep_dir=PREP,
    res_dir=RES,
    pred_dir=OUT_DIR,
    trainer=TRAINER,
    plans=PLANS,
    cfg=CFG,
    fold=FOLD,
    out_examples=OUT_EXAMPLES,
    top_k=TOP_K,
    bottom_k=BOTTOM_K
)
```

```python


import os, time, json, types, sys, inspect
from pathlib import Path
import numpy as np
import nibabel as nib
import torch


RAW = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/dataset/nnUNet_raw")
RES  = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/dataset/nnUNet_results")

TASK    = "Dataset701_MultiTaskSegClass"
CFG     = "3d_fullres"
PLANS   = "nnUNetResEncUNetMPlans"
TRAINER = "nnUNetTrainerChallenge"
FOLD    = 0

IMAGES_TS = RAW / TASK / "imagesTs"

OUT_DIR_TS = RES / TASK / f"{TRAINER}__{PLANS}__{CFG}" / "imagesTs_pred_fast"
DISABLE_TTA = True

stub_name = "nnunetv2.training.nnUNetTrainer.fast_predict_and_eval"
if stub_name not in sys.modules:
    sys.modules[stub_name] = types.ModuleType(stub_name)


from nnunetv2.inference.predict_from_raw_data import nnUNetPredictor
from nnunetv2.imageio.simpleitk_reader_writer import SimpleITKIO


def _amp_dtype():
    if torch.cuda.is_available():
        if torch.cuda.is_bf16_supported():
            return torch.bfloat16
        return torch.float16
    return torch.float32

def _find_checkpoint(model_root: Path, fold: int) -> str:
    """Return checkpoint filename that exists under model_root/fold_{fold}."""
    fold_dir = model_root / f"fold_{fold}"
    for name in ("checkpoint_final.pth", "checkpoint_best.pth"):
        if (fold_dir / name).exists():
            return name
    cands = sorted(fold_dir.glob("*.pth"))
    if cands:
        return cands[0].name
    raise FileNotFoundError(f"No checkpoint found in {fold_dir}")

def _list_test_case_ids(images_ts_dir: Path):
    files = sorted(images_ts_dir.glob("*.nii.gz"))

    cids = sorted({p.stem.rsplit("_", 1)[0] for p in files})
    return cids

def _case_channel_files(images_ts_dir: Path, cid: str):
    
    files = sorted(images_ts_dir.glob(f"{cid}_*.nii.gz"), key=lambda p: p.name)
    if not files:
        raise FileNotFoundError(f"No modality files for {cid}")
    return [str(p) for p in files]



def build_predictor(model_root: Path, folds=(0,), disable_tta=True, verbose=False):

    torch.backends.cudnn.benchmark = True
    torch.backends.cuda.matmul.allow_tf32 = True
    torch.backends.cudnn.allow_tf32 = True
    torch.set_float32_matmul_precision("high")
    amp_dtype = _amp_dtype()
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    first_fold = int(folds[0]) if isinstance(folds, (list, tuple)) and len(folds) else 0
    ckpt_name = _find_checkpoint(model_root, first_fold)

    predictor = nnUNetPredictor(
        tile_step_size=0.5,
        use_gaussian=True,
        use_mirroring=not disable_tta,
        verbose=verbose,
        verbose_preprocessing=False,
        allow_tqdm=False,
        device=device
    )
    if hasattr(predictor, "perform_everything_on_gpu"):
        try:
            predictor.perform_everything_on_gpu = True
        except Exception:
            pass

    init_fn = predictor.initialize_from_trained_model_folder
    params = list(inspect.signature(init_fn).parameters.keys())

    try:
        if "use_folds" in params:
            init_fn(str(model_root), use_folds=folds, checkpoint_name=ckpt_name)
        elif "folds" in params:
            init_fn(str(model_root), folds=folds, checkpoint_name=ckpt_name)
        else:
            init_fn(str(model_root), checkpoint_name=ckpt_name)
            if hasattr(predictor, "folds"):
                predictor.folds = list(folds)
    except TypeError:
        init_fn(str(model_root), checkpoint_name=ckpt_name)
        if hasattr(predictor, "folds"):
            predictor.folds = list(folds)

    net = predictor.network
    net.eval().to(device)

    for m in net.modules():
        if isinstance(m, torch.nn.Conv3d):
            try:
                m.to(memory_format=torch.channels_last_3d)
            except Exception:
                pass

    return predictor, device, amp_dtype

def predict_testset(images_ts_dir: Path, out_dir: Path, model_root: Path, folds=(0,), disable_tta=True, verbose=False):
    out_dir.mkdir(parents=True, exist_ok=True)
    reader = SimpleITKIO()
    predictor, device, amp_dtype = build_predictor(model_root, folds=folds, disable_tta=disable_tta, verbose=verbose)

    cids = _list_test_case_ids(images_ts_dir)
    print(f"[test] found {len(cids)} cases in imagesTs")

    times = []
    with torch.inference_mode():
        for cid in cids:
            img_list = _case_channel_files(images_ts_dir, cid)
            arrays, props = reader.read_images(img_list)

           
            ref_img = nib.load(img_list[0])
            ref_shape = ref_img.shape
            ref_affine = ref_img.affine

            t0 = time.time()
            with torch.autocast(device_type="cuda", dtype=amp_dtype, enabled=torch.cuda.is_available()):
                try:
                    out = predictor.predict_single_npy_array(arrays, props, seg_from_prev_stage=None)
                except TypeError:
                    out = predictor.predict_single_npy_array(arrays, props)
            pred_label = out[0] if isinstance(out, tuple) else out
            times.append(time.time() - t0)

      
            if pred_label.shape == (ref_shape[2], ref_shape[1], ref_shape[0]):
                pred_to_save = np.transpose(pred_label, (2, 1, 0))
            else:
                pred_to_save = pred_label

            nib.save(nib.Nifti1Image(pred_to_save.astype(np.uint8), ref_affine), str(out_dir / f"{cid}.nii.gz"))

    if times:
        print(f"[test] wrote {len(times)} cases to {out_dir}")
        print(f"[test] mean inference time: {float(np.mean(times)):.2f} s per case")
    else:
        print("[test] no cases processed")


model_root = RES / TASK / f"{TRAINER}__{PLANS}__{CFG}" 
predict_testset(IMAGES_TS, OUT_DIR_TS, model_root, folds=(FOLD,), disable_tta=DISABLE_TTA, verbose=False)
```

```python


import os
from pathlib import Path
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt


RAW = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/dataset/nnUNet_raw")
RES = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/dataset/nnUNet_results")

TASK    = "Dataset701_MultiTaskSegClass"
CFG     = "3d_fullres"
PLANS   = "nnUNetResEncUNetMPlans"
TRAINER = "nnUNetTrainerChallenge"

IMAGES_TS = RAW / TASK / "imagesTs"
PRED_DIR  = RES / TASK / f"{TRAINER}__{PLANS}__{CFG}" / "imagesTs_pred_fast"
OUT_DIR   = RES / TASK / f"{TRAINER}__{PLANS}__{CFG}" / "imagesTs_overlays_png"

OUT_DIR.mkdir(parents=True, exist_ok=True)


def list_case_ids(images_ts_dir: Path):
    files = sorted(images_ts_dir.glob("*.nii.gz"))

    return sorted({p.stem.rsplit("_", 1)[0] for p in files})

def pick_image_file(images_ts_dir: Path, cid: str):

    c0 = images_ts_dir / f"{cid}_0000.nii.gz"
    if c0.exists():
        return c0
    hits = sorted(images_ts_dir.glob(f"{cid}_*.nii.gz"))
    if not hits:
        raise FileNotFoundError(f"No image found for {cid}")
    return hits[0]

def window_uint8(im, p_low=1, p_high=99):
    im = im.astype(np.float32)
    lo, hi = np.percentile(im, [p_low, p_high])
    if hi <= lo:
        lo, hi = float(im.min()), float(im.max())
        if hi <= lo:
            return np.zeros_like(im, dtype=np.uint8)
    im = np.clip((im - lo) / (hi - lo), 0, 1)
    return (im * 255.0 + 0.5).astype(np.uint8)

def ensure_xy_z_alignment(pred, ref_shape):

    if pred.shape == (ref_shape[2], ref_shape[1], ref_shape[0]):
        return np.transpose(pred, (2, 1, 0))
    return pred

def find_best_slice(mask_xyz):
    """
    Prefer max area of label 2 per axial slice.
    If no label 2, use union of label 1 and 2.
    Returns z index and a tuple of 2D masks (m1, m2) for that slice.
    """
    m2 = (mask_xyz == 2)
    areas = m2.reshape((-1, m2.shape[2])).sum(axis=0)  
    if areas.max() > 0:
        z = int(np.argmax(areas))
    else:
        mU = np.logical_or(mask_xyz == 1, mask_xyz == 2)
        areasU = mU.reshape((-1, mU.shape[2])).sum(axis=0)
        z = int(np.argmax(areasU)) if areasU.max() > 0 else mask_xyz.shape[2] // 2
    return z, (mask_xyz[:, :, z] == 1), (mask_xyz[:, :, z] == 2)


import matplotlib as mpl
mpl.rcParams["savefig.facecolor"] = "black"
mpl.rcParams["figure.facecolor"] = "black"
mpl.rcParams["axes.facecolor"] = "black"

def draw_and_save(img2d_uint8, m1_2d, m2_2d, out_png, title=None):
    fig = plt.figure(figsize=(6, 6), facecolor="black")
    ax = plt.axes()
    ax.set_facecolor("black")
    ax.imshow(img2d_uint8, cmap="gray", interpolation="nearest")  
    # contours
    try:
        ax.contour(m1_2d.astype(float), levels=[0.5], colors=["c"], linewidths=2)  # label 1
    except Exception:
        pass
    try:
        ax.contour(m2_2d.astype(float), levels=[0.5], colors=["r"], linewidths=2)  # label 2
    except Exception:
        pass
    if title:
        ax.set_title(title, fontsize=11, color="white")
    ax.axis("off")
    plt.tight_layout(pad=0)
    plt.savefig(out_png, dpi=200, facecolor="black")
    plt.close(fig)


---
cids = list_case_ids(IMAGES_TS)
if not cids:
    raise RuntimeError(f"No cases found in {IMAGES_TS}")

saved = 0
skipped = []

for cid in cids:
    img_path  = pick_image_file(IMAGES_TS, cid)
    pred_path = PRED_DIR / f"{cid}.nii.gz"
    if not pred_path.exists():
        skipped.append((cid, "missing prediction"))
        continue

    try:
        img_nii  = nib.load(str(img_path))
        pred_nii = nib.load(str(pred_path))
        img_arr  = np.asarray(img_nii.dataobj) 
        pred_arr = np.asarray(pred_nii.dataobj)


        pred_arr = ensure_xy_z_alignment(pred_arr, img_arr.shape)
        if pred_arr.shape != img_arr.shape:
            skipped.append((cid, f"shape mismatch {img_arr.shape} vs {pred_arr.shape}"))
            continue

        # pick best slice
        z, m1_2d, m2_2d = find_best_slice(pred_arr)

        
        img2d = img_arr[:, :, z]
        img2d_u8 = window_uint8(img2d, 1, 99)

        out_png = OUT_DIR / f"{cid}_z{z:03d}.png"
        title = f"{cid} | z={z} | label1=cyan, label2=red"
        draw_and_save(img2d_u8, m1_2d, m2_2d, str(out_png), title=title)
        saved += 1
    except Exception as e:
        skipped.append((cid, f"error: {e}"))

print(f"Done. Saved {saved} PNGs to: {OUT_DIR}")
if skipped:
    print("Skipped cases:")
    for cid, why in skipped[:10]:
        print(f"  - {cid}: {why}")
    if len(skipped) > 10:
        print(f"  ... and {len(skipped) - 10} more")
```

```python
import json, csv, re
from pathlib import Path
import numpy as np
import nibabel as nib

# ================= config =================
RES  = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/dataset/nnUNet_results")
PREP = Path("/local/data0/home/naumanb/Codes/13b_nnUnet/dataset/nnUNet_preprocessed")

TASK    = "Dataset701_MultiTaskSegClass"
TRAINER = "nnUNetTrainerChallenge"
PLANS   = "nnUNetResEncUNetMPlans"
CFG     = "3d_fullres"
FOLD    = 0

run_root = RES / TASK / f"{TRAINER}__{PLANS}__{CFG}" / f"fold_{FOLD}"
val_dir  = run_root / "validation"
splits_p = PREP / TASK / "splits_final.json"

# ================= helpers =================
CID_RE = re.compile(r"^(quiz)_(\d)_(\d+)$")

def norm_cid(s: str) -> str:
    return s.replace(".nii.gz","").replace(".nii","").strip()

def cls_from_cid(cid: str):
    m = CID_RE.match(cid)
    return int(m.group(2)) if m else None

def infer_class_from_mask(mask_arr: np.ndarray) -> int:
    m = mask_arr.astype(np.int16, copy=False)
    if np.any(m == 2):
        return 2
    if np.any(m == 1):
        return 1
    return 0

def precision_recall_f1(cm: np.ndarray):
    C = cm.shape[0]; eps = 1e-12
    p = np.zeros(C); r = np.zeros(C); f1 = np.zeros(C)
    for c in range(C):
        tp = cm[c, c]
        fp = cm[:, c].sum() - tp
        fn = cm[c, :].sum() - tp
        p[c]  = tp / (tp + fp + eps)
        r[c]  = tp / (tp + fn + eps)
        f1[c] = 2 * tp / (2 * tp + fp + fn + eps)
    return p, r, f1


with open(splits_p) as f:
    split = json.load(f)[0]
val_ids = [norm_cid(x) for x in split["val"]]
print("Validation IDs:", len(val_ids))


present = {norm_cid(p.name) for p in val_dir.glob("*.nii.gz")}
case_ids = [cid for cid in val_ids if cid in present]
print("Found predictions for:", len(case_ids))


for cid in case_ids:
    c = cls_from_cid(cid)
    if c is not None:
        gt[cid] = c

if not gt:
    raise RuntimeError("Could not parse any GT labels from IDs like quiz_<c>_<id>.")

C = max(gt.values()) + 1
print("Detected classes:", sorted(set(gt.values())), "| C =", C)


rows = []
y_true, y_pred = [], []
missing = []

for cid in case_ids:
    f = val_dir / f"{cid}.nii.gz"
    if not f.exists():
        missing.append(cid)
        continue
    mask = nib.load(str(f)).get_fdata()
    if mask.dtype.kind == "f":
        mask = np.rint(mask)
    yhat = infer_class_from_mask(mask)
    ytru = gt[cid]
    rows.append([cid, int(ytru), int(yhat)])
    y_true.append(int(ytru))
    y_pred.append(int(yhat))

if missing:
    print(f"Missing prediction files: {len(missing)}; sample:", missing[:5])

print("Evaluated cases:", len(rows))
if not rows:
    raise RuntimeError("No overlapping GT and prediction cases.")


cm = np.zeros((C, C), dtype=np.int64)
for t, p in zip(y_true, y_pred):
    if 0 <= t < C and 0 <= p < C:
        cm[t, p] += 1

overall_acc = float(np.trace(cm) / np.sum(cm))
row_counts = cm.sum(axis=1)
per_class_acc = np.divide(np.diag(cm), row_counts, out=np.zeros_like(row_counts, dtype=float), where=row_counts > 0)

prec, rec, f1 = precision_recall_f1(cm)
macro_f1 = float(np.mean(f1))

print(f"Overall accuracy: {overall_acc:.4f}")
for i, a in enumerate(per_class_acc):
    print(f"Per-class accuracy {i}: {a:.4f}")
for i in range(C):
    print(f"Class {i}: P {prec[i]:.4f}  R {rec[i]:.4f}  F1 {f1[i]:.4f}")
print("Macro F1:", f"{macro_f1:.4f}")
print("Confusion matrix (rows=true, cols=pred):")
print(cm)


out_dir = val_dir / "cls_eval_from_masks"
out_dir.mkdir(parents=True, exist_ok=True)

per_case_csv = out_dir / "val_cls_per_case.csv"
with open(per_case_csv, "w", newline="") as f:
    w = csv.writer(f)
    w.writerow(["case_id", "y_true", "y_pred"])
    w.writerows(rows)

summary_csv = out_dir / "val_cls_summary.csv"
with open(summary_csv, "w", newline="") as f:
    w = csv.writer(f)
    w.writerow(["metric", "value"])
    w.writerow(["overall_accuracy", f"{overall_acc:.6f}"])
    for i, a in enumerate(per_class_acc):
        w.writerow([f"per_class_accuracy_{i}", f"{a:.6f}"])
    w.writerow(["macro_f1", f"{macro_f1:.6f}"])
    for i in range(C):
        w.writerow([f"precision_{i}", f"{prec[i]:.6f}"])
        w.writerow([f"recall_{i}", f"{rec[i]:.6f}"])
        w.writerow([f"f1_{i}", f"{f1[i]:.6f}"])

cm_csv = out_dir / "val_cls_confusion_matrix.csv"
np.savetxt(cm_csv, cm, fmt="%d", delimiter=",")

print("Per-case CSV:", per_case_csv)
print("Summary CSV:", summary_csv)
print("Confusion matrix CSV:", cm_csv)
```
